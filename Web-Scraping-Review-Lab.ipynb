## Scrape www.ibm.com


Import the required modules and functions


```python
from bs4 import BeautifulSoup   # used to parse HTML
import requests                 # used to download web pages
```

Download the contents of the webpage


```python
# Website we are scraping
url = "https://www.ibm.com"
```

```python
# Download the page content
data = requests.get(url).text
```

Create a soup object using the class BeautifulSoup


```python
# Create the BeautifulSoup object
soup = BeautifulSoup(data, "html.parser")
```

Scrape all links


```python
# Grab every <a> tag (anchors = links) on the page
for link in soup.find_all('a'):
    # Each <a> usually has an href="..." which is the actual URL
    print(link.get('href'))
```

Scrape  all images


```python
for link in soup.find_all('img'):# in html image is represented by the tag <img>
    print(link.get('src'))
```

## Scrape data from html tables


```python
# This page contains a table with color names and their hex codes
URL = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBM-DA0321EN-SkillsNetwork/labs/datasets/HTMLColorCodes.html"
```

Before proceeding to scrape a website, you need to examine the contents, and the way data is organized on the website. Open the above URL in your browser and check how many rows and columns are there in the color table.


```python
# Download the webpage
data = requests.get(URL).text
```

```python
# Turn it into a BeautifulSoup object so we can scrape it
soup = BeautifulSoup(data, "html.parser")
```

```python
# Find the first <table> tag on the page
# (This is the big color code table we want)
table = soup.find('table')
table  # show the raw table HTML so we know we grabbed the right thing
```

# Get all rows from the table


```python
for row in table.find_all('tr'): # in html table row is represented by the tag <tr>
    # Get all columns in each row.
    cols = row.find_all('td') # in html a column is represented by the tag <td>
    color_name = cols[2].getText() # store the value in column 3 as color_name
    color_code = cols[3].getText() # store the value in column 4 as color_code
    print("{}--->{}".format(color_name,color_code))
```
